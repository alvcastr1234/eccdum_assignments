```python
# Importamos librerias

import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import torch.backends.cudnn as cudnn
import numpy as np
import torchvision
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import time
import os
from PIL import Image
from tempfile import TemporaryDirectory

cudnn.benchmark = True
plt.ion()   # interactive mode
```

```python
import os
from torchvision import datasets, transforms
import torch

# Ruta de la carpeta GroceryStoreDataset en Colab
data_dir = '/content/GroceryStoreDataset/newdataset/'

# Aplicamos las transformaciones a los conjuntos de datos de 'train', 'val' y 'test'
data_transforms = {
    'train': transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'test': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}

# Se crean los conjuntos de datos de 'train', 'val' y 'test' usando ImageFolder
image_datasets = {x: datasets.ImageFolder(os.path.join(data_dir, x),
                                          data_transforms[x])
                  for x in ['train', 'val', 'test']}

# Se definen los DataLoaders para 'train', 'val' y 'test'
dataloaders = {x: torch.utils.data.DataLoader(image_datasets[x], batch_size=4,
                                             shuffle=True, num_workers=4)
              for x in ['train', 'val', 'test']}

# Obtenemos el tamaño de los datasets
dataset_sizes = {x: len(image_datasets[x]) for x in ['train', 'val', 'test']}
class_names = image_datasets['train'].classes

# Definimos el dispositivo (GPU si está disponible)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
def imshow(inp, title=None):
    """Display image for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(dataloaders['train']))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
#ENTRENAR EL MODELO
def train_model(model, criterion, optimizer, scheduler, num_epochs=25):
    since = time.time()

    # Create a temporary directory to save training checkpoints
    with TemporaryDirectory() as tempdir:
        best_model_params_path = os.path.join(tempdir, 'best_model_params.pt')

        torch.save(model.state_dict(), best_model_params_path)
        best_acc = 0.0

        for epoch in range(num_epochs):
            print(f'Epoch {epoch}/{num_epochs - 1}')
            print('-' * 10)

            # Each epoch has a training and validation phase
            for phase in ['train', 'val']:
                if phase == 'train':
                    model.train()  # Set model to training mode
                else:
                    model.eval()   # Set model to evaluate mode

                running_loss = 0.0
                running_corrects = 0

                # Iterate over data.
                for inputs, labels in dataloaders[phase]:
                    inputs = inputs.to(device)
                    labels = labels.to(device)

                    # zero the parameter gradients
                    optimizer.zero_grad()

                    # forward
                    # track history if only in train
                    with torch.set_grad_enabled(phase == 'train'):
                        outputs = model(inputs)
                        _, preds = torch.max(outputs, 1)
                        loss = criterion(outputs, labels)

                        # backward + optimize only if in training phase
                        if phase == 'train':
                            loss.backward()
                            optimizer.step()

                    # statistics
                    running_loss += loss.item() * inputs.size(0)
                    running_corrects += torch.sum(preds == labels.data)
                if phase == 'train':
                    scheduler.step()

                epoch_loss = running_loss / dataset_sizes[phase]
                epoch_acc = running_corrects.double() / dataset_sizes[phase]

                print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

                # deep copy the model
                if phase == 'val' and epoch_acc > best_acc:
                    best_acc = epoch_acc
                    torch.save(model.state_dict(), best_model_params_path)

            print()

        time_elapsed = time.time() - since
        print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
        print(f'Best val Acc: {best_acc:4f}')

        # load best model weights
        model.load_state_dict(torch.load(best_model_params_path, weights_only=True))
    return model
```

```python
def visualize_model(model, num_images=6):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(dataloaders['val']):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images // 2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'Predicted: {class_names[preds[j]]}\nActual: {class_names[labels[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
model_ft = models.resnet18(weights='IMAGENET1K_V1')
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 5.
# Alternatively, it can be generalized to ``nn.Linear(num_ftrs, len(class_names))``.
model_ft.fc = nn.Linear(num_ftrs, 5)

model_ft = model_ft.to(device)

criterion = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer_ft = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)

# Decay LR by a factor of 0.1 every 7 epochs
exp_lr_scheduler = lr_scheduler.StepLR(optimizer_ft, step_size=7, gamma=0.1)
```

```python
model_ft = train_model(model_ft, criterion, optimizer_ft, exp_lr_scheduler, num_epochs=25)
```

```python
visualize_model(model_ft)
```

```python
# Diccionario de precios de los productos
precios = {
    'Apple': 74,
    'Pear': 89,
    'Satsumas': 44,
    'Melon': 109,
    'Orange': 35
}

# Función para calcular el costo/riesgo para cada predicción
def calcular_costo(y_pred, y_real):
    if y_pred > y_real:
        return 5 * (y_pred - y_real)  # El usuario paga de más
    elif y_pred < y_real:
        return y_real - y_pred  # El usuario paga de menos
    else:
        return 0  # Escenario ideal

# Evaluar el costo total en el conjunto de validación
costos_totales = []
for inputs, labels in dataloaders['test']:
    inputs = inputs.to(device)
    labels = labels.to(device)

    outputs = model_ft(inputs)
    _, preds = torch.max(outputs, 1)

    for pred, label in zip(preds, labels):
        producto_predicho = class_names[pred]
        producto_real = class_names[label]

        y_pred = precios[producto_predicho]
        y_real = precios[producto_real]

        costo = calcular_costo(y_pred, y_real)
        costos_totales.append(costo)

# Calcular el costo/riesgo total y promedio
costo_total = sum(costos_totales)
costo_promedio = costo_total / len(costos_totales)

print(f'Costo total para el cliente: ${costo_total:.2f}')
print(f'Costo promedio por predicción: ${costo_promedio:.2f}')
```

```python
# Lista para almacenar los errores
errores = []

for inputs, labels in dataloaders['test']:
    inputs = inputs.to(device)
    labels = labels.to(device)

    outputs = model_ft(inputs)
    _, preds = torch.max(outputs, 1)

    for i in range(len(preds)):
        producto_predicho = class_names[preds[i]]
        producto_real = class_names[labels[i]]

        y_pred = precios[producto_predicho]
        y_real = precios[producto_real]

        # Calcular el costo
        costo = calcular_costo(y_pred, y_real)

        # Verificar si hay un error y almacenar la información
        if producto_predicho != producto_real:
            errores.append({
                'Producto Real': producto_real,
                'Producto Predicho': producto_predicho,
                'Costo': costo
            })

# Mostrar los errores y costos asociados
import pandas as pd

df_errores = pd.DataFrame(errores)
print("Resumen de errores:")
print(df_errores)

# Mostrar la tabla de errores en el notebook
from IPython.display import display

display(df_errores)
```

```python
# Contar el número total de productos evaluados en el conjunto de validación
cantidad_total = len(df_errores)  # Número total de predicciones erróneas registradas

# Contar la cantidad total de productos evaluados
total_evaluados = sum(dataset_sizes.values())  # Tamaño total de los datasets de 'train' y 'val'

print(f'Cantidad total de productos evaluados: {total_evaluados}')
print(f'Cantidad total de productos con error: {cantidad_total}')
```

```python
# Creamos diccionarios para errores y costos por categoría
errores_por_categoria = {class_name: 0 for class_name in class_names}
costos_por_categoria = {class_name: 0 for class_name in class_names}

# Procesamos el DataFrame de errores
for _, row in df_errores.iterrows():
    producto_real = row['Producto Real']
    costo = row['Costo']

    # Incrementamos errores y costos para la categoría real
    errores_por_categoria[producto_real] += 1
    costos_por_categoria[producto_real] += costo

# Recalculamos métricas de performance basadas en etiquetas reales y predichas
from sklearn.metrics import classification_report

true_labels = []
pred_labels = []

for inputs, labels in dataloaders['test']:
    inputs = inputs.to(device)
    labels = labels.to(device)

    outputs = model_ft(inputs)
    _, preds = torch.max(outputs, 1)

    true_labels.extend(labels.cpu().numpy())
    pred_labels.extend(preds.cpu().numpy())

# Generamos el reporte de clasificación
report = classification_report(true_labels, pred_labels, target_names=class_names, output_dict=True)

# Creamos DataFrame resumen
df_resumen = pd.DataFrame({
    'Categoría': class_names,
    'Precisión': [report[cat]['precision'] for cat in class_names],
    'Recall': [report[cat]['recall'] for cat in class_names],
    'F1-Score': [report[cat]['f1-score'] for cat in class_names],
    'Soporte': [report[cat]['support'] for cat in class_names],
    'Errores': [errores_por_categoria[categoria] for categoria in class_names],
    'Costo Total': [costos_por_categoria[categoria] for categoria in class_names],
    'Costo Promedio': [
        costos_por_categoria[categoria] / errores_por_categoria[categoria]
        if errores_por_categoria[categoria] > 0 else 0
        for categoria in class_names
    ]
})

# Mostramos los resultados
print("\nResumen de errores, costos y performance por categoría:")
from IPython.display import display
display(df_resumen)

# Calculamos totales generales
total_errores = sum(errores_por_categoria.values())
total_costos = sum(costos_por_categoria.values())

print(f"\nTotal de errores: {total_errores}")
print(f"Total de costos: {total_costos}")
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix

# Generamos la matriz de confusión
cm = confusion_matrix(true_labels, pred_labels, labels=range(len(class_names)))

# Convertimos la matriz de confusión en un DataFrame con etiquetas
df_cm = pd.DataFrame(cm, index=class_names, columns=class_names)

# Mostramos la matriz de confusión con etiquetas
print("Matriz de confusión con etiquetas:")
from IPython.display import display
display(df_cm)
```

```python
# Calculamos métricas (Precisión y Recall) para cada categoría
precisions = []
recalls = []

for i, class_name in enumerate(class_names):
    true_positives = cm[i, i]
    false_positives = sum(cm[:, i]) - true_positives
    false_negatives = sum(cm[i, :]) - true_positives

    # Precisión
    if true_positives + false_positives > 0:
        precision = true_positives / (true_positives + false_positives)
    else:
        precision = 0

    # Recall
    if true_positives + false_negatives > 0:
        recall = true_positives / (true_positives + false_negatives)
    else:
        recall = 0

    # Agregamos a las listas
    precisions.append(precision)
    recalls.append(recall)

    # Imprimimos métricas para cada categoría
    print(f"{class_name}:")
    print(f"  Precisión: {precision:.3f}")
    print(f"  Recall: {recall:.3f}\n")

# Creamosr un DataFrame resumen
df_metrics = pd.DataFrame({
    "Categoría": class_names,
    "Precisión": precisions,
    "Recall": recalls
})

# Mostramos el DataFrame
print("\nResumen de métricas por categoría:")
from IPython.display import display
display(df_metrics)
```

## Estrategias Comerciales Derivadas del Análisis

Basándonos en el desempeño y los errores de las distintas categorías, se proponen las siguientes estrategias comerciales para mitigar los problemas detectados y potenciar las oportunidades identificadas.

**1. Reducción del Impacto de los Errores Costosos**
- Las categorías Apple y Orange representan un desafío debido a los altos costos asociados a los errores. Algunas acciones recomendadas incluyen:
  - **Ajuste de Margen**: Incrementar ligeramente los márgenes de estas categorías para absorber el impacto económico.
  - **Promociones de Productos Relacionados**: Incentivar productos asociados para aumentar el ticket promedio y compensar los costos de error. Por ejemplo:
    - "Compra 2 kg de Orange y obtén un descuento en jugos de naranja."

**2. Promoción y Reposicionamiento de Satsumas**
- Dado el bajo rendimiento de Satsumas, es clave impulsar esta categoría mediante:
  - **Promociones Especiales**: Ofertas que capten la atención de los clientes, como:
    - "Llévate 2 kg de Satsumas al precio de 1."
  - **Diferenciación en Estanterías**: Separar Satsumas de productos similares como Orange para reducir confusiones.
  - **Campañas Educativas**: Informar a los clientes sobre las características únicas de Satsumas (sabor, origen, beneficios) mediante materiales en el punto de venta.

**3. Minimizar la Confusión entre Categorías**
- La confusión entre Orange y Satsumas es un problema recurrente. Se recomienda:
  - **Reorganización del Punto de Venta**: Ubicar los productos con suficiente distancia entre ellos y utilizar señalización clara.
  - **Destacar las Diferencias**: Usar etiquetas o empaques que resalten las características únicas de cada producto.

**4. Aprovechamiento de las Categorías Mejor Performantes**
- Las categorías con buen desempeño (Melon y Apple) pueden ser usadas estratégicamente:
  - **Campañas de Posicionamiento**:
    - Ejemplo: "El melón más fresco, listo para disfrutar en cualquier momento."
  - **Ofertas Combinadas**:
    - Ejemplo: "Compra 1 kg de Apple y recibe un 20% de descuento en Melon."
  - **Gestión de Inventario**: Asegurar siempre la disponibilidad de estos productos para capitalizar su popularidad.

**5. Ajuste de Precios Basado en Costos**
- Las categorías con errores más costosos (Apple y Orange) pueden beneficiarse de:
  - **Precios Dinámicos**: Ajustar precios según la temporada y la demanda para maximizar los márgenes.
  - **Redistribución de Margen**: Compensar los costos de errores con márgenes ajustados en otras categorías menos afectadas.

**6. Comunicación y Confianza del Cliente**
- Para mitigar el impacto de los errores en la percepción de los clientes:
  - **Política de Garantías**: Asegurar que los clientes puedan devolver o cambiar productos en caso de insatisfacción.
    - Ejemplo: "¿No estás satisfecho con tu compra? Cambia tu producto sin costo."
  - **Transparencia en los Esfuerzos**: Informar sobre las acciones que la empresa está tomando para mejorar la calidad y precisión de la clasificación.
