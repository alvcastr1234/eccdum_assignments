## Unifying Victoria's Secret

```python
def unify_victoria_secret(df):
    """
    We want that all brands that are related to Victoria's Secret
    have `victoria's secret` as their brand instead of what they
    currently have.
    """
    df = df.copy()
    new_string = "victoria's secret"

    # Lista de patrones o variaciones que queremos reemplazar
    patterns = [
        r"victoria'?s?-?\s?secret.*",  # Captura 'victorias-secret', 'victoria secret', etc.
        r"victorias-secret"  # Captura la variación sin espacio ni apóstrofe
    ]

    # Aplicamos el reemplazo en la columna 'brand_name'
    for pattern in patterns:
        df['brand_name'] = df['brand_name'].str.lower().str.replace(pattern, new_string, regex=True)

    return df

# Aplicamos la función y verificamos los resultados
df_unified = unify_victoria_secret(df)
```

## Cleaning up the price

In this sectino we are going to transform the `price` column into a float column in USD dolars.

For this, be careful of the different formats in the data.

For simplicity, you might assume that all the prices are in USD dolars, regarding of the symbol of the currency used.

```python
import re

def clean_price(df):
    """
    In this function we will transform the
    `price` column into a column of floats.
    In case a product has more than one price,
    return the lowest one.
    """
    df = df.copy()  # Evitar modificar el DataFrame original

    def extract_min_price(price_str):
        """Extrae el precio más bajo de una cadena."""
        if not isinstance(price_str, str) or price_str.strip() == "":
            return None  # Manejar valores vacíos o no strings

        # Usamos una expresión regular para encontrar precios con o sin decimales
        prices = re.findall(r'\d+\.?\d*', price_str)  # Captura precios enteros y decimales
        prices = [float(price) for price in prices]

        if prices:
            return min(prices)  # Devolvemos el menor precio si hay alguno válido
        return None  # Si no hay precios, devolvemos None

    # Aplicamos la función a la columna 'price'
    df['price'] = df['price'].apply(extract_min_price)

    # Eliminamos las filas con precios faltantes
    df = df.dropna(subset=['price'])

    return df
```

```python
df_clean = clean_price(df_unified)
```

```python
def low_high_product_mean(df):
    """
    Finally, we will calculate `product_category` with the lowest and highest mean price
    for the brand Victoria's Secret.
    """
    # Filtramos los productos solo de la marca 'victoria's secret'
    vs_products = df[df['brand_name'] == "victoria's secret"]

    # Calculamos la media de los precios agrupados por 'product_category'
    category_means = vs_products.groupby('product_category')['price'].mean()

    # Encontramos la categoría con la media más baja y más alta
    lowest_mean = category_means.min()
    highest_mean = category_means.max()

    return lowest_mean, highest_mean
```

```python
lowest_mean, highest_mean = low_high_product_mean(df_clean)
assert np.allclose(lowest_mean, 3.6203030303030)
print("highest_mean", highest_mean)
```
